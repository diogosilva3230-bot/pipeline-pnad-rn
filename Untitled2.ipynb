```python
base_completa_rn = []

anos = [2023, 2024, 2025]

print("Iniciando automação inteligente...\n")

for ano in anos:

    url_base = f"https://ftp.ibge.gov.br/Trabalho_e_Rendimento/Pesquisa_Nacional_por_Amostra_de_Domicilios_continua/Trimestral/Microdados/{ano}/"

    try:
        resposta = requests.get(url_base)
        soup = BeautifulSoup(resposta.text, "html.parser")

        # pegar todos os arquivos .zip disponíveis
        arquivos = [a.get('href') for a in soup.find_all('a') if a.get('href').endswith('.zip')]

        print(f"\n📂 Ano {ano} - {len(arquivos)} arquivos encontrados")

        for arquivo in arquivos:
            try:
                url_arquivo = url_base + arquivo
                print(f"⬇️ Baixando: {arquivo}")

                r = requests.get(url_arquivo)

                with zipfile.ZipFile(io.BytesIO(r.content)) as z:

                    txt_nome = [nome for nome in z.namelist() if nome.endswith('.txt')][0]

                    with z.open(txt_nome) as f:

                        df = pd.read_fwf(f, colspecs=posicoes, names=nomes_colunas)

                        # filtro RN (UF = 24)
                        df['UF'] = df['UF'].astype(str).str.zfill(2)
                        df_rn = df[df['UF'] == "24"]

                        base_completa_rn.append(df_rn)

                        print(f"✅ Processado: {arquivo}")

            except Exception as e:
                print(f"❌ Erro no arquivo {arquivo}: {e}")

    except Exception as e:
        print(f"❌ Erro ao acessar o ano {ano}: {e}")

# ==============================
# CONSOLIDAÇÃO FINAL
# ==============================

if base_completa_rn:
    df_final = pd.concat(base_completa_rn, ignore_index=True)

    df_final.to_csv("PNADC_RN_2023_2025.csv", index=False)

    print("\n--- PIPELINE FINALIZADO COM SUCESSO ---")
    print(f"Total de registros do RN: {len(df_final)}")
    print(df_final.head())

else:
    print("❌ Nenhum dado foi coletado.")

    print("\n--- PIPELINE FINALIZADO COM SUCESSO ---")
print(f"Total de registros do RN: {len(df_final)}")

print("\n📊 Primeiras linhas dos dados:")
print(df_final.head())

print("\n📊 Registros por ano:")
print(df_final['Ano'].value_counts())

print("\n📊 Registros por trimestre:")
print(df_final['Trimestre'].value_counts())

print("\n📊 Cruzamento Ano x Trimestre:")
print(pd.crosstab(df_final['Ano'], df_final['Trimestre']))

```
